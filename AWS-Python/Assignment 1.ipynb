```python
#Create Variables with any Object using Various Naming Conventions

#Integer
x = 7
print(x)

#Float
y = 3.14
print(y)

#String
message = "Hello Welcome to AWS python course"
print(message)

#Boolean
boolean_true = True
boolean_false = False

#List
list = [1,2,3,'a','b','c']
#Assecing elements of a list
print("First Element in list: ", list[0])
#List slicing
print("Sliced list: ",list[1:5])

#Tuple
my_tuple = (1,2,3,'x','y','z')
#Assecing elements of the tuple
print("First Element in tuple: ", my_tuple[2])
#Tuple Length
print("Length of tuple :", len(my_tuple))

#Snake Case
snakecase = "this_is_a_sentence_in_snake_case"
print("Snake Case:",snakecase)
#Pascal Case
pascelcase = "ThisIsASentenceInPascalCase"
print("Pascel Case:",pascelcase)
#Camel Case
camelcase = "thisIsASentenceInCamelCase"
print("Camel Case:",camelcase)
#Upper Case with underscore
uppercase_with_underscore = "THIS_IS_A_SENTENCE_IN_UPPERCASE_WITH_UNDERSCORE"
print("Upper Case with underscore:",uppercase_with_underscore)
```

```python
#Reverse a String using slicing 
string = "Welcome to the course. All the best"
reversed_string = string[::-1]
print(reversed_string)
```
